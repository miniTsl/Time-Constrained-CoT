```python
from matplotlib import pyplot as plt
import os
import numpy as np
import json

file_dir = "../outputs/12_11/"
model_list = ["Qwen/Qwen2.5-7B-Instruct", "Qwen/Qwen2.5-3B-Instruct", "Qwen/Qwen2.5-1.5B-Instruct", "Qwen/Qwen2.5-0.5B-Instruct"]
```

```python
dataset = "gsm8k"

prompt_type1 = "qwen25-math-cot"
prompt_type2 = "coarse-to-fine-structured"

something = "_-1_seed0_t0.0_s0_e-1_"
acc_1 = []
acc_2 = []
for model in model_list:
    with open(os.path.join(file_dir, model, prompt_type1, dataset, "test_" + prompt_type1 + something + prompt_type1 + "_metrics.json")) as f:
        data1 = json.load(f)
        acc_1.append(data1["acc"])
    with open(os.path.join(file_dir, model, prompt_type2, dataset, "test_" + prompt_type2 + something + prompt_type2 + "_metrics.json")) as f:
        data2 = json.load(f)
        acc_2.append(data2["acc"])
```

```python
# plot pillar bars for each model, and wrtie the accuracy on top of the bars
barWidth = 0.25
r1 = np.arange(len(model_list))
r2 = [x + barWidth for x in r1]

plt.bar(r1, acc_1, color='b', width=barWidth, edgecolor='grey', label=prompt_type1)
plt.bar(r2, acc_2, color='r', width=barWidth, edgecolor='grey', label=prompt_type2)

# 在柱子顶部添加数值标签
for i in range(len(model_list)):
    plt.text(r1[i], acc_1[i], f'{acc_1[i]}', ha='center', va='bottom')
    plt.text(r2[i], acc_2[i], f'{acc_2[i]}', ha='center', va='bottom')

plt.xlabel('Models', fontweight='bold')
plt.xticks([r + barWidth/2 for r in range(len(model_list))], ["7B", "3B", "1.5B", "0.5B"])
plt.ylabel('Accuracy')
plt.title('Accuracy on ' + dataset)
plt.legend()
plt.show()

```
